
### Influence of height


### Study done for x-axis on mt1

```python

d = df[df["height"]==1]["1-x-mean"]
d = d.reset_index(drop=True)
d.plot(label="h=1",legend=True)
d = df[df["height"]==2]["1-x-mean"]
d = d.reset_index(drop=True)
d.plot(label="h=2",legend=True)
d = df[df["height"]==3]["1-x-mean"]
d = d.reset_index(drop=True)
d.plot(label="h=3",legend=True)
```

### Study done for x-axis average on all probes

```python
def getSum(df,condition,columnLabels):
    return sum([df[eval(condition)][label] for label in columnLabels])/len(columnLabels)
    
cols = [f"{i}-x-mean" for i in range(1,5)]
d = getSum(df,"df['height']==1",cols)

d = getSum(df,"df['height']==1",cols)
d = d.reset_index(drop=True)
d.plot(label="h=1",legend=True)
d = getSum(df,"df['height']==2",cols)
d = d.reset_index(drop=True)
d.plot(label="h=2",legend=True)
d = getSum(df,"df['height']==3",cols)
d = d.reset_index(drop=True)
d.plot(label="h=3",legend=True)
```

### Study done for y-axis average on all probes

```python
cols = [f"{i}-y-mean" for i in range(1,5)]
d = getSum(df,"df['height']==1",cols)
d = d.reset_index(drop=True)
d.plot(label="h=1",legend=True)
d = getSum(df,"df['height']==2",cols)
d = d.reset_index(drop=True)
d.plot(label="h=2",legend=True)
d = getSum(df,"df['height']==3",cols)
d = d.reset_index(drop=True)
d.plot(label="h=3",legend=True)
```

### Study done for z-axis average on all probes

```python
cols = [f"{i}-z-mean" for i in range(1,5)]
d = getSum(df,"df['height']==1",cols)
d = d.reset_index(drop=True)
d.plot(label="h=1",legend=True)
d = getSum(df,"df['height']==2",cols)
d = d.reset_index(drop=True)
d.plot(label="h=2",legend=True)
d = getSum(df,"df['height']==3",cols)
d = d.reset_index(drop=True)
d.plot(label="h=3",legend=True)
```

## Influence of lid and mockup on shocks

Reported as function of the height

```python
cols = [f"{i}-x-mean" for i in range(1,5)]
d = getSum(df,"df['lid']==False",cols)
d = d.reset_index(drop=True)
d.plot(label="w/o lid - empty",legend=True)
d = getSum(df,"(df['lid']==1) & (df['mockup']==0)",cols)
d = d.reset_index(drop=True)
d.plot(label="w lid - empty",legend=True)
d = getSum(df,"(df['lid']==1) & (df['mockup']==1)",cols)
d = d.reset_index(drop=True)
d.plot(label="w lid - filled",legend=True)
```
